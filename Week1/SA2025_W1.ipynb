### 1. What is name of car that has the highest horsepower?  

```python
max_hp_row = cars.loc[cars['horsepower'].idxmax()]
car_name = max_hp_row['name']
print(f"Car with maximum horsepower: {car_name} with {max_hp_row.horsepower} hp")
```

### 2. How many cars have mpg  35?

```python
high_mpg_cars = cars[cars['mpg'] >= 35]
count = len(high_mpg_cars)
print(f"Number of cars with mpg  35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = cars[(cars['horsepower'] > 100) & (cars['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print(f"Most common origin for cars with horsepower > 100 and weight < 3000: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japanese_cars = cars[cars['origin'] == 'japan']
mean_acc = round(japanese_cars['acceleration'].mean(), 2)
print(f"Mean acceleration of cars from Japan: {mean_acc}")
```

### 5. Which year had the highest average mpg?

```python
avg_mpg_per_year = cars.groupby('model_year')['mpg'].mean()
year_with_max_avg_mpg = avg_mpg_per_year.idxmax()
max_avg_mpg = round(avg_mpg_per_year.max(), 2)
print(f"Year with highest average mpg: {year_with_max_avg_mpg} (Average mpg: {max_avg_mpg})")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = cars['mpg'].median()

filtered = cars[cars['mpg'] > median_mpg].copy()

filtered['hp_weight_ratio'] = filtered['horsepower'] / filtered['weight']

max_ratio = filtered['hp_weight_ratio'].max()

best_cars = filtered[filtered['hp_weight_ratio'] == max_ratio]

print(f"Car(s) with best horsepower-to-weight ratio (among those with above-median mpg): {best_cars['name'].values}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg = cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over Years')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
grouped = cars.groupby('name').agg(
    year_count=('model_year', pd.Series.nunique),
    appearances=('name', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
).reset_index()

consistent_models = grouped[(grouped['year_count'] > 1) & (grouped['mpg_std'] < 1.0)].copy()

consistent_models['mpg_mean'] = consistent_models['mpg_mean'].round(2)
consistent_models['mpg_std'] = consistent_models['mpg_std'].round(3)

consistent_models_sorted = consistent_models.sort_values(
    by=['appearances', 'mpg_mean'], ascending=[False, False]
)

result = consistent_models_sorted[['name', 'appearances', 'mpg_mean']]

print("Consistent car models (low mpg variation, multiple years):")
print(result)
```
